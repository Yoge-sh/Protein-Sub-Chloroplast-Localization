```python
def create_word2vec_embeddings(data, model_W2V):
    embeddings = []
    sentences = data
    for sentence in sentences:
        vectors = [model_W2V.wv[word] for word in sentence if word in model_W2V.wv]
        if vectors:
            doc_vector = np.mean(vectors, axis=0)
        else:
            doc_vector = np.zeros(model_W2V.vector_size)
        embeddings.append(doc_vector)
    return embeddings
```

```python
def create_fasttext_embeddings(data, model_FT):
    embeddings = []
    sentences = data
    for sentence in sentences:
        vectors = [model_FT.wv[word] for word in sentence if word in model_FT.wv]
        if vectors:
            doc_vector = np.mean(vectors, axis=0)
        else:
            doc_vector = np.zeros(model_FT.vector_size)
        embeddings.append(doc_vector)

    return embeddings
```

## SkipXgram for PSSM and HMM profile

```python
# Skipgram based on PSSM or HMM
def SXGbg(arr, X):
    res = np.zeros((20,20))
    L = len(arr)
    # arr = arr[0]
    # print(arr[0])
    for i in range(20):
        for j in range(20):
            for k in range(L):
                # print(arr[k,i], arr[(k+X+1) % L,j])
                res[i,j] += arr[k,i] * arr[(k+X+1) % L,j]

    res.resize(1,400)
    return res
```

## SkipXGram for Protein Sequence

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.models import Sequential
```

```python
# Skipgram based on protien sequence
def skipxgram_bigrams(input_string,x):
    skip_1_gram_bigrams = []
    for i in range(len(input_string) - x):
        bigram = input_string[i] + input_string[i + x]
        skip_1_gram_bigrams.append(bigram)
    bigram_counts = Counter(skip_1_gram_bigrams)
    vector_length = 400
    frequency_vector = [0] * vector_length
    for i in range(vector_length):
        bigram = chr(65 + (i // 20)) + chr(65 + (i % 20))
        frequency_vector[i] = bigram_counts.get(bigram, 0)

    return frequency_vector
```

```python
# # Skipgram based on protien sequence
final_df['skip1gram'] = final_df['Sequence_x'].apply(lambda x: skipxgram_bigrams(x,1))
final_df['skip2gram'] = final_df['Sequence_x'].apply(lambda x: skipxgram_bigrams(x,2))
final_df['skip3gram'] = final_df['Sequence_x'].apply(lambda x: skipxgram_bigrams(x,3))
final_df['skip4gram'] = final_df['Sequence_x'].apply(lambda x: skipxgram_bigrams(x,4))
final_df['skip5gram'] = final_df['Sequence_x'].apply(lambda x: skipxgram_bigrams(x,5))
```

```python
from sklearn.model_selection import train_test_split

# Converting numpy array to tensor for using in tf model
def tensor_split(X, y, test_size):
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=test_size, random_state=101)
    tensor_list = [tf.convert_to_tensor(np_arr, dtype=tf.float32) for np_arr in X_train]

    # Create a TensorFlow tensor containing all the tensors
    X_train = tf.stack(tensor_list)

    tensor_list2 = [tf.convert_to_tensor(np_arr, dtype=tf.float32) for np_arr in X_test]

    # Create a TensorFlow tensor containing all the tensors
    X_test = tf.stack(tensor_list2)
    return X_train, X_test, Y_train, Y_test
```

## NN model

```python
from tensorflow.python.keras.models import Model,Input
from tensorflow.keras.layers import LSTM, Embedding, Dense, TimeDistributed, Dropout, Bidirectional
```

```python
# defining the model, will tune later
def train_nn(X, y, X_train, X_test, Y_train, Y_test, epochs, name):
  l1 = len(X_train[0])
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(l1, activation='relu'),
    tf.keras.layers.Dense(l1//10, activation='relu'),
    tf.keras.layers.Dense(l1//50, activation='relu'),
    tf.keras.layers.Dense(l1//100, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
  ])
  # input = Input(shape=(2000,))
  # model = Embedding(input_dim=462, output_dim=2000, input_length=2000)(input)
  # # model = Dropout(0.1)(model)
  # model = Bidirectional(LSTM(units=100, return_sequences=True))(input)#(model)
  # out = TimeDistributed(Dense(1, activation="softmax"))(model)  # softmax output layer

  # model = Model(input, out)
  model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy']) ## acuracy
  early_stopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy', ## acc
                                                        patience=10)
  model.fit(X_train, Y_train, batch_size=32, epochs=epochs, callbacks=[early_stopping])

  train_loss, train_acc = model.evaluate(X_train,  Y_train)
  test_loss, test_acc = model.evaluate(X_test,  Y_test)#, verbose=2)
  #
  tensor_list = [tf.convert_to_tensor(np_arr, dtype=tf.float32) for np_arr in X]
  X_relax = tf.stack(tensor_list)
  relax_loss, relax_acc = model.evaluate(X_relax,  y, verbose=2)

  print('\nTrain accuracy:', train_acc)
  print('\nTest accuracy:', test_acc)
  print('\nRelax accuracy:', relax_acc)

  #Mine1
  pred1 = model.predict(X_test) ##
  df_test = pd.DataFrame(pred1,columns=['pred_{}'.format(name)])
  #Mine2
  pred2 = model.predict(X_relax)
  df_relax = pd.DataFrame(pred2,columns=['pred_{}'.format(name)])
  def biout(x):
    if x >= 0.5:
      return 1
    else:
      return 0
  df_test['pred_{}'.format(name)] = df_test['pred_{}'.format(name)].apply(biout)
  df_relax['pred_{}'.format(name)] = df_relax['pred_{}'.format(name)].apply(biout)
  return model, df_test, df_relax
```
